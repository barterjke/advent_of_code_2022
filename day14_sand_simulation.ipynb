```python
import copy

with open("15_1.txt") as file:
    content = file.read()
arr = [[[int(a) for a in b.split(",")] for b in c.split(" -> ")] for c in content.split("\n")]
```

```python
from itertools import chain

chained = list(chain.from_iterable(arr)) + [(500, 0)]
min_p = min(chained, key=lambda x: x[0])[0], min(chained, key=lambda x: x[1])[1]
max_p = max(chained, key=lambda x: x[0])[0], max(chained, key=lambda x: x[1])[1]
```

```python
y_range = max_p[1] + 3 - min_p[1]
x_range = max(max_p[0] + 2 - (min_p[0] - 1), 2 * y_range)
grid = [[0 for j in range(y_range)] for i in range(x_range)]
```

```python
from copy import deepcopy
from numpy import sign

min_x, min_y = min_p
offset = x_range // 2 - 500
grid[500 + offset][min_y] = 2
for line in deepcopy(arr):
    for i, section in enumerate(line):
        line[i] = section[0] + offset, section[1] - min_y
        if i == 0:
            continue
        p1, p2 = line[i], line[i - 1]
        dif_x, dif_y = sign(p2[0] - p1[0]), sign(p2[1] - p1[1])
        while p1 != p2:
            x, y = p1
            grid[x][y] = 1
            p1 = x + dif_x, y + dif_y
        grid[p2[0]][p2[1]] = 1
for i in range(x_range):
    grid[i][max_p[1] + 2] = 1
```

```python
add = lambda x, y: (x[0] + y[0], x[1] + y[1])
```

```python
init_pos = 500 + offset, min_y

def spawn(grid):
    cur_pos = init_pos
    moved = True
    while moved:
        moved = False
        for _dir in [(0, 1), (-1, 1), (1, 1)]:
            next_x, next_y = add(cur_pos, _dir)
            if grid[next_x][next_y] == 0:
                moved = True
                cur_pos = next_x, next_y
                break
    x, y = cur_pos
    grid[x][y] = 3
    return cur_pos == init_pos
```

```python
def draw_grid(grid):
    for y in range(len(grid[0])):
        for x in range(len(grid)):
            print({0: '.', 1: "#", 2: "+", 3: 'o'}.get(grid[x][y]), end = "")
        print()
```

```python
from IPython.display import clear_output
import time

def looped_spawn(grid, to_draw=True, sec=1):
    while True:
        if spawn(grid):
            break
        clear_output()
        to_draw and draw_grid(grid)
        if to_draw:
            time.sleep(sec)
```

```python
try:
    grid = copy.deepcopy(grid)
    looped_spawn(grid, False, 0.1)
except IndexError:
    pass
clear_output()
draw_grid(grid)
r = sum([len(list(filter(lambda y: y == 3, x))) for x in grid])
print(r)
```
